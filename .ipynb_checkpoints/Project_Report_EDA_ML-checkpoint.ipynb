```python
# Since multiple charts are required for analyisis, trying to create  generic function to plot charts which will be re-used

def charts(df,col,title,xrotation,order_list):
    sns.set_style("darkgrid")
    plt.figure(figsize=(14,8))
    if order_list == None:
        ax =sns.countplot(df[col], hue =df['default_prediction'], palette='crest',color='matplotlib color')
    else:
        ax =sns.countplot(df[col], hue =df['default_prediction'], palette='crest', order = order_list)
    plt.title(title, fontsize=18, fontweight='bold')
    plt.xticks(fontsize=14,rotation = xrotation)
    plt.yticks(fontsize=14)
    plt.xlabel(col,fontsize=16)
    plt.ylabel("Count",fontsize=16)
    plt.legend(fontsize=16)
    total = float(len(df['default_prediction']) )
    for p in ax.patches:
            percentage = '{:1.2f}%'.format(100 * p.get_height()/total)
            x = p.get_x() + p.get_width()/1.8
            y = p.get_height()+ 20
            ax.annotate(percentage, (x, y),ha='center')  
    plt.show()   
```

```python
# defining linearsvm model
def svm_model(X_train, y_train, X_test, y_test):
    """
        Description:
        Implementation of SVM Model
        :param X_train: Feature Training Data
        :param Y_train:  Prediction Training Data
        :param x_test:  Feature Testing Data
        :param y_test:  Prediction Testing Data
        :return: Predicted data set
    """

    svm_classifier = svm.LinearSVC()
    svm_classifier.fit(X_train, y_train)
    y_pred = svm_classifier.predict(X_test)
    print("SVM Train Accuracy " + str(svm_classifier.score(X_train, y_train)*100) + "%")
    print("SVM Test Accuracy " + str(svm_classifier.score(X_test, y_test)*100) + "%\n")
    print("Classification Report for SVM Model")
    print(classification_report(y_test, y_pred))
    return y_pred
```

##### Although above model definition & execution works fine. Creating generic functions so that multiple model execution is performed by calling them directly

```python
# Defining model execution function

def models_exe (X_train, y_train, X_test, y_test, models):
    """ Generic function to run different model and check accuracy
    
    Args:
         X_train(pd.DataFrame): train dataset 
         y_train(pd.DataFrame): train predict dataset
         X_test(pd.DataFrame): test dataset
         y_test(pd.DataFrame): test predict dataset
    """
    
    
    dfs = []
     
    # List of classifier model to be tested
    models = models
    print(models)
    
    results = []
    
    names = []
    
    scoring = ['accuracy', 'precision_weighted', 'recall_weighted', 'f1_weighted', 'roc_auc']
    
    target_names = ['0','1']
    
    all_cf_matrix = []
    
    for name, model in models:
        start_time = time.time()
        kfold = model_selection.KFold(n_splits=5, shuffle=True, random_state=77)
        cv_results = model_selection.cross_validate(model, X_train, y_train, cv=kfold, scoring=scoring)
        clf = model.fit(X_train, y_train)
        y_test_pred = clf.predict(X_test)
        print ('-'*50)
        print ('{0}:'.format(name))
        print ('Classification Matrix:')
        print(classification_report(y_test, y_test_pred, target_names=target_names))
        print ('Confusion Matrix:')
        cf_matrix = confusion_matrix(y_test, y_test_pred)
        plot_confusion_matrix(cf_matrix,name)
        print("Time taken: {:0.2f} seconds".format(time.time() - start_time))
        results.append(cv_results)
        names.append(name)
        this_df = pd.DataFrame(cv_results)
        this_df['model'] = name
        dfs.append(this_df)        
        final = pd.concat(dfs, ignore_index=True)    
        
    return final
```

```python
# Generic function to plot confustion matrix for different models
def plot_confusion_matrix(cf_matrix,title):
    plt.figure(figsize=(6.4,4.8))
    group_names = ['True Neg','False Pos','False Neg','True Pos']

    group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]

    group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]

    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]

    labels = np.asarray(labels).reshape(2,2)

    ax = sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')

    ax.set_title(f'Confusion Matrix for accuracy of {title}\n\n');
    ax.set_xlabel('\nPredicted Default')
    ax.set_ylabel('Actual Default ');

    ax.xaxis.set_ticklabels(['0','1'])
    ax.yaxis.set_ticklabels(['0','1'])

    plt.show()
```

```python
# List of classifier model to be tested
models = [ ('Logistic Regression', LogisticRegression(random_state=77)), 
               ('Random Forest', RandomForestClassifier(random_state=77)),
               ('K-Nearest Neighbours', KNeighborsClassifier()),
               ('Support Vector Machines', SVC(random_state=77)), 
               ('Gaussian Naive Bayes', GaussianNB()),
               #('Naive Bayes', MultinomialNB()),
               ('Decision Tree', DecisionTreeClassifier(random_state=77)),
               ('Bagging Decision Tree', BaggingClassifier(DecisionTreeClassifier(random_state=77),max_samples=0.5,
                                   max_features=1.0,n_estimators=10)),
               ('Boosting Decision Tree',AdaBoostClassifier(DecisionTreeClassifier(min_samples_split=10,max_depth=4,
                                   random_state=77),n_estimators=10,learning_rate=0.6))   
             ] 

# Running model and getting results
result = models_exe(X_train,y_train,X_test,y_test,models)
```

```python
# importing module
import re

# text file sample

Text_file="""
profile 1: 
Number of emails: 7 
sender_email: hello@gmail.com
sender_name: "Mr. Aman"
recipient_email: R@new.com
recipient_name: None
date_sent: 31 Oct 2020
subject: Assistance
email_body: email body here

profile 2:
Number of emails: 2
sender_email: sheena@gmail.com
sender_name: "Mrs. Sheena"
recipient_email: New@pathfinder.com
recipient_name: Nitin
date_sent: 1 Nov 2019
subject: new product feature
email_body: email body here
"""

#pattern search 
pattern=r"\w\S*@gmail.com"

# find the email id's (gmail.com) from the text file

re.findall(pattern, Text_file)
```
